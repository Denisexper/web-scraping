```python
from bs4 import BeautifulSoup as bs
import requests
import csv

url = "https://portal.siges.sv/pp/sedes"
resultado = requests.get(url)
html = resultado.text
soup = bs(html, 'html.parser')

box = soup.find('ul', id='form:basicDT_list')
nombres = box.find_all('span', class_='tituloitemSedeCEO')
nombres_list = []


for item in nombres:
    # buscamos las etiquetas que contengan el titulo la direccion 
    titulo = item.get_text(strip=True, separator=' ')
    codigo_tag = item.find_next('div', class_='spacecode space')
    direccion_tag = codigo_tag.find_next('div', class_='space')
    contacto_tag = direccion_tag.find_next('div', class_='space')

    # agregamos la imformacion de las etiquetas a la lista
    escuela_info = {
        'Titulo': titulo,
        'Codigo': codigo_tag.get_text(strip=True) if codigo_tag else 'No se encontró código',
        'Direccion': direccion_tag.get_text(strip=True) if direccion_tag else 'No se encontró dirección',
        'Contacto': contacto_tag.get_text(strip=True) if contacto_tag else 'No se encontró contacto'
    }

    # con esta imprimimos la imfromacion en la terminal
    print("=" * 80)
    print(escuela_info['Titulo'])
    print(escuela_info['Codigo'])
    print(escuela_info['Direccion'])
    print(escuela_info['Contacto'])

    # agrega la imformacion de las escuelas a la lista que creamos
    nombres_list.append(escuela_info)

# Guarda la información en un archivo CSV
with open('escuelas.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Titulo', 'Codigo', 'Direccion', 'Contacto']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(nombres_list)
```
